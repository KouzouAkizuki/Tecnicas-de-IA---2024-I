**1. Estudiar el uso de Redes Neuronales en scikit-learn, haciendo énfasis en la función MLPClassifier**

**a. Machine Learning con redes neuronales usando scikit-learn y su func**

```python
# Paso 1 - Carga de librerías y módulos necesarios
# Librerías
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import sklearn
from sklearn.neural_network import MLPClassifier
from sklearn.neural_network import MLPRegressor
# Módulos
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, classification_report, confusion_matrix
from math import sqrt
```

```python
# Paso 2 - Lectura de datos y comprobación
df = pd.read_csv('diabetes.csv')
print(df.shape)
df.describe().transpose()
```

```python
# Paso 3 - Creación de arreglo de características y variable de respuesta
target_column = ['diabetes']

predictors = list(set(list(df.columns))-set(target_column))  #Lista con todos los encabezados menos 'diabetes'
df[predictors] = df[predictors]/df[predictors].max()
df.describe().transpose()
```

```python
# Paso 4 - Creación de los conjuntos de entrenamiento y prueba

X = df[predictors].values
y = df[target_column].values

# 30% datos de prueba y 70% datos de entrenamiento
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 40)
print(X_train.shape)
print(X_test.shape)
```

```python
# Paso 5 - Construcción, Predicción, y evaluación del modelo de red neuronal
# Se utiliza la función de activación lineal rectificada y un limite de 500 iteraciones

mlp = MLPClassifier(hidden_layer_sizes = (8,8,8), activation = 'relu', solver = 'adam', max_iter = 500)     #Red neuronal con 3 capas de neuronas ocultas con 8 neuronas en cada una
mlp.fit(X_train,y_train)
predict_train = mlp.predict(X_train)
predict_test = mlp.predict(X_test)
```

```python
# Resultados de entrenamiento
print(confusion_matrix(y_train,predict_train))
print(classification_report(y_train,predict_train))
```

```python
# Resultados de prueba
print(confusion_matrix(y_test,predict_test))
print(classification_report(y_test,predict_test))
```

Las 2 medidas relevantes es la precisión y la Puntuación F1, esta ultima vide el desempeño entre la precisión y la sensibilidad (recall), lo ideal es que estás dos sean 1. Sin embargo, es difícil de obtener en escenarios del mundo real, por ahora la red neuronal formada por un perceptron multi capas (MLP), presenta un desempeño superior al 70%, sin embargo, presenta un mejor comportamiento en los casos negativos de diabetes.

```python
# Lectura de datos y comprobación
df = pd.read_csv('Project_Data_Fingers.csv')
print(df.shape)
df.describe().transpose()
```

```python
# Creación de arreglo de características y variable de respuesta
target_column = ['Clase']

predictors = list(set(list(df.columns))-set(target_column))  #Lista con todos los encabezados menos 'diabetes'
df[predictors] = df[predictors]/df[predictors].max()
df.describe().transpose()
```

```python
# Creación de los conjuntos de entrenamiento y prueba

X = df[predictors].values
y = df[target_column].values

# 30% datos de prueba y 70% datos de entrenamiento
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.50, random_state = 40)
print(X_train.shape)
print(X_test.shape)
```

```python
# Construcción, Predicción, y evaluación del modelo de red neuronal
# Se utiliza la función de activación lineal rectificada y un limite de 500 iteraciones

mlp = MLPClassifier(hidden_layer_sizes = (29,29,29), activation = 'relu', solver = 'adam', max_iter = 1000)
mlp.fit(X_train,y_train)
predict_train = mlp.predict(X_train)
predict_test = mlp.predict(X_test)
```

```python
# Resultados de entrenamiento
print(confusion_matrix(y_train,predict_train))
print(classification_report(y_train,predict_train))
```

```python
# Resultados de prueba
print(confusion_matrix(y_test,predict_test))
print(classification_report(y_test,predict_test))
```
